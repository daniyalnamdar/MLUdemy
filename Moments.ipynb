# Moments: Mean, Variance, Skew, Kurtosis

```python
import numpy as np 
import matplotlib.pyplot as plt

vals = np.random.normal(4, 15, 10000)

plt.hist(vals, 50)
plt.show()
```

# Mean

```python
np.mean(vals)
```

# variance

```python
np.var(vals)
```

# skew

```python
import scipy.stats as sp
sp.skew(vals)
```

# kurtosis

```python
sp.kurtosis(vals)
```
