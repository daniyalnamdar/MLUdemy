# MatPlotLib Basics

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-3, 3, 0.001)

plt.plot(x, norm.pdf(x))
plt.show()
```

# Plot Multiple plots on one Graph

```python
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1.0, 0.5))
plt.show()
```

# save it to File

```python
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1.0, 0.5))
plt.savefig('c:\\hjgjgjhgj', format='png')
```
